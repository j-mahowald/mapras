```python
import re

import matplotlib.pyplot as plt

s = '''(94.645, 12.53)
(73.824, 9.33)
(48.937, 6.75)
(24.394, 4.58)'''

matches = re.findall(r'\(\s*([0-9]*\.?[0-9]+)\s*,\s*([0-9]*\.?[0-9]+)\s*\)', s)
x = [float(a) for a, b in matches]
y = [float(b) for a, b in matches]

plt.figure(figsize=(8, 4.5))
plt.plot(x, y, marker='o', linestyle='-', color='orangered', markersize=6, linewidth=3)
plt.scatter(x, y)
for xi, yi in zip(x, y):
    plt.annotate(f'{yi}', (xi, yi), textcoords="offset points", xytext=(0,5), ha='center')
plt.xlim(0, 100)
plt.xlabel('Number of embeddings (thousands)')
plt.ylabel('Inference time (seconds)')
plt.title('Inference time vs. number of embeddings')
plt.grid(True)
plt.savefig('inf_time.png', dpi=300)
plt.show()
```
